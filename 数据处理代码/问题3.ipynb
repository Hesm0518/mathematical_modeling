```python
import os

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

import numpy as np
```

```python
# 处理一些冲突
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 设置字体
plt.rcParams["axes.unicode_minus"] = False  # 该语句解决图像中的“-”负号的乱码问题
```

```python
# 数据读取
soil_numb = pd.read_excel('./dataset/14-soil-carbaon.xls')# 土壤数据
soil_numb = soil_numb.round({'SOC土壤有机碳':2,'SIC土壤无机碳':2,'STC土壤全碳':2,'全氮N':2,'土壤C/N比':2})
soil_numb
```

```python
soil_numb_ng17 = soil_numb[(soil_numb['放牧强度（intensity）'] == 'NG')&(soil_numb['放牧小区（plot）']== 'G17')]
soil_numb_ng17.groupby('year').agg({'SOC土壤有机碳': 'mean','SIC土壤无机碳': 'mean','STC土壤全碳': 'mean','全氮N': 'mean','土壤C/N比': 'mean'}).round({'SOC土壤有机碳':2,'SIC土壤无机碳':2,'STC土壤全碳':2,'全氮N':2,'土壤C/N比':2}).reset_index()
```

```python
tableng17 = pd.pivot_table(soil_numb_ng17, values=['SOC土壤有机碳','SIC土壤无机碳','STC土壤全碳','全氮N','土壤C/N比'], columns='year')
tableng17 
```

```python
tableng17 = pd.pivot_table(soil_numb_ng17, values=['SOC土壤有机碳','SIC土壤无机碳','STC土壤全碳','全氮N','土壤C/N比'], columns='year')
sns.lineplot(data=tableng17)
plt.title('G17小区在NG放牧强度下每年的土壤元素变化情况')
plt.ylabel('各土壤元素的数量')
plt.savefig('./G17NG.png')
```
