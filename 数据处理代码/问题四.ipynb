```python
import os

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

import numpy as np
```

```python
# 处理一些冲突
plt.rcParams["font.sans-serif"] = ["SimHei"]  # 设置字体
plt.rcParams["axes.unicode_minus"] = False  # 该语句解决图像中的“-”负号的乱码问题
```

```python
# 数据读取
water_volume = pd.read_excel('./dataset/9-water-volume.xlsx') # 径流量数据
water_volume
```

```python
plt.figure(figsize=(10,8))
tablewatersec = pd.pivot_table(water_volume, values='径流量(m3/s)', index=['月份'], columns=['年份'])

sns.lineplot(data=tablewatersec)
plt.title('各个年份不同月份径流量(m3/s)变化折线图')
plt.ylabel('径流量(m3/s)')
plt.savefig('./Watersec.png')
```

```python
plt.figure(figsize=(10,8))
tablewater = pd.pivot_table(water_volume, values='径流量(m3)', index=['月份'], columns=['年份'])

sns.lineplot(data=tablewater)
plt.title('各个年份不同月份径流量(m3)变化折线图')
plt.ylabel('径流量(m3)')
plt.savefig('./Water.png')
```
